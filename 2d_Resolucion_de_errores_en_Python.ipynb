## Lectura y comprensión de mensajes de error

Es fundamental desarrollar habilidades de lectura y comprensión de errores para poder solucionarlos de manera eficiente. Hay algunos pasos que se pueden seguir para optimizar la resolución de problemas en el código, por lo que nos ayudaremos de un ejemplo de error para reconocer la estructura:




````python
x = "Ximena"

#No se puede sumar un string.
suma = x + 1
print(suma)
````

3. Determinar la línea en la que se encuentra el error para posteriormente corregirlo. Al inicio de cada línea podemos encontrar su número respectivo, adicionalmente Python destaca la línea que presenta el error con una flecha. En este caso el error se encuentra en la ```línea 4```.
4. Si, a pesar de seguir estos pasos, no se logra solucionar el problema, se puede buscar el mensaje de error en distintos foros en línea.


### Ejemplos de errores de Python

Los errores más comunes son:

1. `SyntaxError`: Este error ocurre cuando se encuentra un error de sintaxis en el código, una estructura mal escrita o un paréntesis que no se encuentre cerrado.
2. `NameError`: Se produce al momento de ocupar una variable que no ha sido definida previamente.
3. `TypeError`: Ocurre cuando los tipos de datos no son los correctos.
4. `IndexError`: Ocurre cuando se ocupa un índice fuera del rango existente.
5. `ZeroDivisionError`: Ocurre cuando se divide un número por 0, que matemáticamente es considerada una indefinición.


## Forzado de errores para comprender la situación

1. **SyntaxError**

````python
x = 5
y = 7

if (x > y):
  print(x, "es mayor que ", y)
#Lo correcto sería else
els:
  print(x, "es menor que ", y)
````

_Error corregido_

````python
x = 5
y = 7

if (x > y):
  print(x, "es mayor que ", y)
else:
  print(x, "es menor que ", y)
````

Otro ejemplo de **SyntaxError**, que difiere del anterior por el mensaje es el siguiente.

````python
x = 5
y = 7

if (x > y):
  #No está cerrado el paréntesis.
  print(x, "es mayor que ", y
````

_Error corregido_

````python
x = 5
y = 7

if (x > y):
  print(x, "es mayor que ", y)
````

2. **NameError**

````python
nombre = "Julio"
#La variable edad no está definida.
print(edad)
````

_Error corregido_

````python
nombre = "Julio"
edad = 20
print(edad)
````

3. **TypeError**

````python
x = "Ximena"

#No se puede sumar un string.
suma = x + 1
print(suma)
````

_Error corregido_

````python
x = 5

suma = x + 1
print(suma)
````

4. **IndexError**

````python
ejemplo = (1, 2, 3, 4, 5)

#Índices existentes del 0 al 4.
print(ejemplo[8])
````

_Error corregido_

````python
ejemplo = (1, 2, 3, 4, 5)

print(ejemplo[0])
````

5. **ZeroDivisionError**

````python
a = 0

resultado = 7 / a
print(resultado)
````

_Error corregido_

````python
a = 0.01

resultado = 7/a
print(resultado)
````
