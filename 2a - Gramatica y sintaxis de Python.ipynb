# Gramática y sintaxis de Python

La gramática y la sintaxis de Python son el conjunto de reglas que definen la estructura y el orden de las palabras y símbolos que se utilizan para escribir programas en Python. 

La gramática de Python es muy clara y sencilla de seguir, y se basa en un conjunto de reglas bien definidas. De hecho, la guía de estilo de Python es bien conocida y está disponible en el documento [PEP8](https://peps.python.org/pep-0008/).

Algunas de las características más importantes de la gramática y sintaxis de Python son las siguientes:

## Variables

En Python, se pueden asignar valores a variables utilizando el operador de asignación (=). Por ejemplo:

````python
numero = 10   # asigna el valor 10 a la variable número
texto = "hola"   # asigna el valor "hola" a la variable texto
````

Para definir el nombre de variables se debe considerar los siguientes puntos:

*   El nombre de una variable, función, clase, módulo u objeto. Habitualmente, los identificadores empiezan con letras en minúscula, o con un guion bajo seguido de números o letras. 
*   Se debe considerar que el nombre no puede empezar por un número, no se permite el uso de guiones "-", acentos, ni el uso de espacios.

A continuación se muestran unos ejemplos de nombres de variables válidos y no válidos:

````python
# Válidos

_2_grupo = 10
valor_maximo = 60
grupo_2 = 30
variable = 40
````

````python
# No Válidos

2_variable = 10
numero-mayor = 20
numero menor = 30
````

Adicionalmente, se puede asignar el mismo valor a diferentes variables con el siguiente código.

```

```

O también podemos asignar varios valores separados por coma.

```

```

## Indentación

Indentación: en Python, la indentación es muy importante y se utiliza para definir bloques de código. Los bloques de código se indentan utilizando cuatro espacios en blanco o una tabulación. Por ejemplo:

````python
# Ejemplo de indentación
numero = -10
if numero < 0:  # si x es menor que 0
    print("numero es negativo")  # este código está indentado y forma parte del bloque if
    numero = -numero  # este código también está indentado y forma parte del bloque if
print("Valor final de numero es:", numero)  # este código no está indentado y no forma parte del bloque if
````

## Operadores

Python tiene un conjunto de operadores aritméticos y lógicos que se pueden utilizar para realizar cálculos y comparaciones en el código.

### Operadores aritméticos
Python tiene un conjunto de operadores aritméticos que se pueden utilizar para realizar cálculos matemáticos. Algunos ejemplos de operadores aritméticos son:

````python
a = 10
b = 5

c = a + b  # c contiene el resultado de a + b (15)
d = a - b  # d contiene el resultado de a - b (5)
e = a * b  # e contiene el resultado de a * b (50)
f = a / b  # f contiene el resultado de a / b (2.0)
````

### Operadores de comparación o lógicos

Python tiene un conjunto de operadores de comparación que se pueden utilizar para comparar dos valores y obtener un resultado booleano (verdadero o falso). Algunos ejemplos de operadores de comparación son:

````python
x = 10
y = 20

resultado = x < y  # el resultado de x < y (verdadero)
resultado = x > y  # el resultado de x > y (falso)
resultado = x == y  # el resultado de x == y (falso)
resultado = x != y  # el resultado de x != y (verdadero)
````

### Múltiples líneas

PEP8 recomienda que las líneas no excedan los 79 caracteres. Haciendo uso de \ se puede distribuir el código en varias líneas, lo que en determinados casos hace que el código sea mucho más legible.

```
x = 1 + 2 + 3 + 4 +\
    5 + 3 + 6 + 9
```

Si por lo contrario estamos dentro de un bloque rodeado con paréntesis (), bastaría con saltar a la siguiente línea.

```
x = (1 + 2 + 3 + 4 +
     5 + 3 + 6 + 9)
```

## **1. Reglas de tipado en Python**
Python es un lenguaje de tipado **dinámico** y **fuerte**.

Es de **tipado dinámico**, ya que una variable puede comenzar teniendo un tipo de dato y puede cambiar en cualquier momento a otro tipo de dato. Por ejemplo:

````python
variable_dinamica = 10
print(variable_dinamica)
variable_dinamica = "Hola mundo"
print (variable_dinamica)
````

Como se ha podido observar anteriormente, la variable_dinamica es creada con el valor de 10, que es un número entero (int). Luego, en la tercera línea se asigna el nuevo valor "Hola mundo", por lo cual el tipo de dato cambia a una cadena (str).

Por otro lado, Python es un lenguaje de **tipado fuerte**; ya que ante una operación entre dos tipos de datos incompatibles, arroja un error (durante la compilación o la ejecución) en lugar de convertir implícitamente alguno de los dos tipos. Por ejemplo:

````python
a = 10
b = "5"
print (a + b)
````

## **2. Profundización en las estructuras básicas**

Python cuenta con bucles iterativos como `for` y `while`; sin embargo, en otros lenguajes de programación existen otros como `do while`.

Entonces, ¿cuáles son sus diferencias? ¿Cuándo debería usar uno en lugar de otro?

### FOR
Se usa cuando ya se conoce el número de iteraciones, es decir, cuando se sabe cuántas veces se debe ejecutar una instrucción . Es por eso que se usa con objetos finitos como lo es una lista, un diccionario, un set, etc.



### WHILE
En cambio, el bucle `while` se usa cuando el número de iteraciones puede variar dependiendo de dos factores:


*   De los valores que se “comparan” de manera inicial, pudiendo ser que debido a la condición no se lleguen a ejecutar.
*   Las operaciones que se realicen dentro del mismo ciclo.

Con `while`, mientras la condición sea cierta se ejecuta el cuerpo del bucle. Si la condición es falsa termina el bucle. Solo ejecuta su contenido si se comprueba una condición, lo que puede ocurrir 0, 1 o más veces.

### DO WHILE
Específicamente no se cuenta con el bucle `do while` en Python, pero se puede construir uniendo un bucle `while` y una condición.

Al igual que `while`, el bucle `do while` se usa cuando el número de iteraciones varía; sin embargo, la instrucción `do while` permite repetir una instrucción hasta que una expresión especificada sea falsa. El código se ejecuta al menos una vez aunque la condición sea falsa, porque la condición es verificada después de la ejecución del código.



### Ejemplo de conversión de while / do while / for para el factorial de un número:

#### FOR

````python
numero = 4
factorial = 1
for i in range (1,numero + 1):
    factorial = factorial * i
print ("El factorial del número",numero,"es:",factorial)
````

#### WHILE

````python
numero = 4
i = 0
factorial = 1
while i < numero:
    factorial = factorial * (i+1)
    i = i + 1
print ("El factorial del número ", numero,"es:",factorial)
````

#### DO WHILE

````python
numero = 4
i = 0
factorial = 1
while True:
    factorial = factorial * (i+1)
    i = i + 1
    if numero <= i:
        break
print ("El factorial del número ", numero,"es:",factorial)
````

## **3. Ámbito de las variables**
El ámbito de una variable es el contexto en el que existe esa variable. Por eso, una variable existe en dicho ámbito a partir del momento en que se crea y deja de existir cuando desaparece su ámbito. Los principales tipos de ámbitos en Python son dos: ámbito local y ámbito global.

Para entender mejor los tipos de ámbitos lo haremos mediante funciones; recordemos que las funciones en Python constituyen unidades lógicas de un programa y tienen un doble objetivo:


### Variables globales
Las variables globales son todas las que se crean fuera de una función. Estas variables pueden ser utilizadas en todo el código, tanto dentro como fuera de las funciones.

Ejemplo:

````python
variable_global = "global"

def funcion():
    print("Esta es una variable "+ variable_global)

funcion()
````

### Función
El ámbito de una variable dentro de una función solo existe dentro de la función, estas variables son conocidas como variables locales o de ámbito local. Hay que tener en cuenta que si la variable local tiene el mismo nombre de una variable global, el valor de esta última no se modificará.

Ejemplo:

````python
variable = "global"

def funcion():
    variable = "local"
    print("Esta es una variable",variable)

funcion()
print("Esta es una variable",variable)
````

Se pueden importar/invocar variables globales desde otros módulos o archivos haciendo uso adecuado del import.

### Paquete
El ámbito de las variables de un paquete existirá dentro del paquete y dentro del código en el que se realice la llamada/importación del paquete.

## **4. Paso por valor y paso por referencia**

Dependiendo del tipo de dato que se envíe a la función, se puede diferenciar dos comportamientos: 
* Paso por valor: se crea una copia local de la variable dentro de la función. 
* Paso por referencia: se maneja directamente la variable, y los cambios realizados dentro de la función le afectarán también fuera. A continuación se explicará mas detalladamente:

### **Paso por valor**
Cuando se utiliza paso por valor de los argumentos, lo que se hace es copiar el valor de las variables en los respectivos parámetros. Entonces, cualquier modificación del valor del parámetro no afecta a la variable externa correspondiente. Generalmente, los tipos simples se pasan por valor: enteros, flotantes, cadenas, lógicos, etc.

Ejemplo:

````python
def doble_numero(numero):
    numero = numero * 2

num = 10
doble_numero(num)
print(num)
````

### **Paso por referencia**
Cuando se utiliza paso por referencia, lo que se hace es copiar en los parámetros la dirección de memoria de las variables que se usan como argumento. Esto implica que realmente hagan referencia al mismo objeto/elemento y cualquier modificación del valor en el parámetro afectará a la variable externa correspondiente. Generalmente los tipos compuestos se pasan por referencia: listas, diccionarios, conjuntos, etc.

Ejemplo:

````python
def doble_numero(numeros):
    for i,n in enumerate(numeros):
        numeros[i] *= 2

n = [10,50,100]
doble_numero(n)
print(n)
````
